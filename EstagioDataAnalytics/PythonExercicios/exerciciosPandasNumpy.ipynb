```python
'''Faça um gráfico de barras contendo os top 5 apps por número de instalação.'''
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("googleplaystore.csv")

df.dropna(inplace=True)
df["Installs"] = df["Installs"].str.replace("[,+]","").to_frame()
df["Installs"] = pd.to_numeric(df["Installs"])

newdf = df.groupby(["App"],as_index=False)["Installs"].sum().sort_values(by="Installs",ascending=False).head(5)
newdf.plot(x="App", y="Installs", kind="bar")
plt.show()
```

```python
'''Faça um gráfico de pizza (pie chart) mostrando as categorias de apps existentes no dataset
de acordo com a frequência em que elas aparecem.'''
import pandas as pd
import matplotlib.pyplot as plt

df.dropna(inplace = True)
df = pd.read_csv("googleplaystore.csv")
newdf = df.groupby(["Category"],as_index=False)["Installs"].count().sort_values(by="Installs",ascending=False).head(5)
newdf.rename(columns={"Installs":"Frequencia"})
newdf["Installs"] = pd.to_numeric(newdf["Installs"])



plt.pie(newdf["Installs"],labels=newdf["Category"],autopct="%1.1f%%")
plt.show()
```

```python
'''Mostre qual o app mais caro existente no dataset'''
import pandas as pd
df = pd.read_csv("googleplaystore.csv")

df.dropna(inplace = True)
for x in df.index:
    if df.loc[x, "Price"] == "Everyone" or df.loc[x, "Type"] == "Free":
        df.drop(x, inplace = True)

df["Price"] = df["Price"].str.replace("[$]","")
df["Price"] = pd.to_numeric(df["Price"])


df[["App","Price"]].sort_values(by="Price",ascending=False).head(1).reset_index(drop=True)
```

```python
'''Mostre qual a vinícola (winery) que possui o maior número de reviews (num_reviews) nos
últimos 20 anos.'''
#importa o df
import pandas as pd
df = pd.read_csv("wines_SPA.csv")

#separa e cria as colunas
df['winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity'].str.split(';', expand=True)
df[['winery', 'wine', 'year','rating','num_reviews', 'country', 'region','price','type', 'body', 'acidity']] = df['winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity'].str.split(';', expand=True)
df.drop(["winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity"],axis='columns',inplace=True)

#limpa os dados
df.dropna(inplace=True)
df.dropna(inplace = True)
for x in df.index:
    if df.loc[x, "year"] == "N.V.":
        df.drop(x, inplace = True)       
df["year"] = pd.to_numeric(df["year"])
for x in df.index:
    if df.loc[x, "year"] < (2022-20):
        df.drop(x, inplace = True)
df["num_reviews"] = pd.to_numeric(df["num_reviews"])

#agrupa vinicolas, soma as reviews, e ordena
newdf = df.groupby(["winery"])["num_reviews"].sum().to_frame()
newdf = newdf.sort_values(by="num_reviews",ascending=False).head(1).reset_index()
newdf.rename(columns={"winery":"Vinícola","num_reviews": "reviewsTotais"},inplace=True)
newdf
```

```python
'''Mostre quais são as 5 regiões com a média de preço mais alta e mais baixa'''
import pandas as pd
df = pd.read_csv("wines_SPA.csv")

df['winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity'].str.split(';', expand=True)
df[['winery', 'wine', 'year','rating','num_reviews', 'country', 'region','price','type', 'body', 'acidity']] = df['winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity'].str.split(';', expand=True)
df.drop(["winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity"],axis='columns',inplace=True)

df.dropna(inplace=True)
for x in df.index:
    if df.loc[x, "price"].replace('.','',1).isdigit() == False:
        df.drop(x, inplace = True)  
df["price"] = pd.to_numeric(df["price"])


newdf = df.groupby(["region"])["price"].mean().to_frame()
newdf = newdf.sort_values(by="price",ascending=False).head(5).reset_index()
newdf.rename(columns={"region":"Região","price": "Preco Medio"},inplace=True)


newdf2 = df.groupby(["region"])["price"].mean().to_frame()
newdf2 = newdf2.sort_values(by="price").head(5).reset_index()
newdf2.rename(columns={"region":"Região","price": "Preco Medio"},inplace=True)


newdf3 = newdf.append(newdf2,ignore_index=True)
newdf3
```

```python
'''A acidez (acidity) varia entre valores até 5, mostre a média de acidez por ano em todo o
intervalo existente na coluna ano (year).'''
import pandas as pd
df = pd.read_csv("wines_SPA.csv")

df['winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity'].str.split(';', expand=True)
df[['winery', 'wine', 'year','rating','num_reviews', 'country', 'region','price','type', 'body', 'acidity']] = df['winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity'].str.split(';', expand=True)
df.drop(["winery;wine;year;rating;num_reviews;country;region;price;type;body;acidity"],axis='columns',inplace=True)

df.dropna(inplace=True)
for x in df.index:
    if df.loc[x, "acidity"] == "NA":
        df.drop(x, inplace = True)
for x in df.index:
    if df.loc[x, "year"] == "" or df.loc[x, "year"] == "NA" or df.loc[x, "year"] == "N.V.":
        df.drop(x, inplace = True)
df["acidity"] = pd.to_numeric(df["acidity"])

pd.set_option('display.max_rows', None)
newdf = df.groupby(["year"])["acidity"].mean().to_frame().reset_index()
newdf["acidity"] = newdf["acidity"].round(decimals=2)
newdf.rename(columns={"year":"Ano","acidity": "Media acidity"},inplace=True)
newdf
```

```python
'''Crie um array 5x5 com a sequência 1...25 (incluso) e faça a soma dos elementos da diagonal.'''
import numpy as np
 
arr = np.arange(1,26)
newarr = np.reshape(arr,(5,5))

print(newarr)
print(np.trace(newarr))
```

```python
'''Crie um array 5x5 com a sequência de números pares entre 0...50 (incluso) e faça as seguintes operações:'''
import numpy as np
 
arr = np.arange(0,50,2)
newarr = np.reshape(arr,(5,5))

print(newarr)

#soma linhas
print("Soma linha 1:",np.sum(newarr[0]),"linha 2:",np.sum(newarr[1]),"linha 3:",np.sum(newarr[2]),"linha 4:",np.sum(newarr[3]),"inha 5:",np.sum(newarr[4]))
#inverte colunas por linha e soma linhas
print("Soma coluna 1:",np.sum(newarr.T[0]),"coluna 2:",np.sum(newarr.T[1]),"coluna 3:",np.sum(newarr.T[2]),"coluna 4:",np.sum(newarr.T[3]),"coluna 5:",np.sum(newarr.T[4]))
#media ultima linha
print("Media ultima linha:",np.average(newarr[-1]))
#inverte colunas por linha e faz media ultima linha
print("Media ultima coluna:",np.average(newarr.T[-1]))
```

```python
'''Dado o array definido por np.arange(25).reshape(5, 5), realize as seguintes operações sem utilizar nenhum
tipo de loop (for / while):'''

arr = np.arange(25).reshape(5, 5)
print(arr)

print("Media linha 0:",np.average(arr[0]),"\nMedia linha 2:",np.average(arr[2]),"\nMedia linha 3:",np.average(arr[3]))
print("Media coluna 0:",np.average(arr.T[0]),"\nMedia coluna 1:",np.average(arr.T[1]),"\nMedia coluna 4:",np.average(arr.T[4]))

print("A soma das duas diagonais:",np.trace(arr) + np.trace(arr.T))
```

```python
import re


f = open("actors.csv", "r")
#abre os arquivos para escrita
e1 = open("exercicioDesafio1.txt", "w")
e2 = open("exercicioDesafio2.txt", "w")
e3 = open("exercicioDesafio3.txt", "w")
e4 = open("exercicioDesafio4.txt", "w")
e5 = open("exercicioDesafio5.txt", "w")


#tranforma todo o conteudo em uma string
string = str(f.read())

#separa as quebras de linhas por virgulas
string = string.replace('\n',',')

#separa toda a string por "," ignorando algumas partes especificadas
PATTERN = re.compile(r'''((?:[^,"]|"[^"]*"|'[^']*')+)''')
stringSplitada = PATTERN.split(string)[1::2]



#criando listas para cada coluna
Actor=[]
Total_Gross=[]
Number_of_Movies=[]
Average_per_Movie=[]
Fisrt_Movie=[]
Gross=[]


#adiciona os elementos de cada coluna em uma lista e converte eles
for i in range(6,len(stringSplitada),6):
    Actor.append(stringSplitada[i])
    
for i in range(7,len(stringSplitada),6):
    Total_Gross.append(stringSplitada[i])
Total_Gross=list(map(float,Total_Gross))  
    
for i in range(8,len(stringSplitada),6):
    Number_of_Movies.append(stringSplitada[i])
Number_of_Movies=list(map(int,Number_of_Movies))
    
for i in range(9,len(stringSplitada),6):
    Average_per_Movie.append(stringSplitada[i])
Average_per_Movie=list(map(float,Average_per_Movie))
    
for i in range(10,len(stringSplitada),6):
    Fisrt_Movie.append(stringSplitada[i])
    
for i in range(11,len(stringSplitada),6):
    Gross.append(stringSplitada[i])
Gross=list(map(float,Gross))

    

#printa as listas separadas em colunas da tabela

#print(Actor)
#print(Total_Gross)
#print(Number_of_Movies)
#print(Average_per_Movie)
#print(Fisrt_Movie)
#print(Gross)


#EXERCICIOS
'''O ator/atriz com maior número de filmes e o respectivo número de filmes.'''
maiorNumeroFilmes = max(Number_of_Movies)
index_maiorNumeroFilmes = Number_of_Movies.index(maiorNumeroFilmes)
e1.write('O ator com maior numero de filmes é: "{}", com {} filmes.\n\n'.format(Actor[index_maiorNumeroFilmes],maiorNumeroFilmes))



'''A média do número de filmes por autor.'''
for i in range(0,len(Actor)):
    e2.write('"{}", media numero de filmes: {}\n'.format(Actor[i],Number_of_Movies[i]))



'''O ator/atriz com a maior média por filme. '''
maiorMediaPorFilme = max(Average_per_Movie)
index_Average_per_Movie = Average_per_Movie.index(maiorMediaPorFilme)
e3.write('\nO ator com maior media por filme é: "{}", com {} de media.\n\n'.format(Actor[index_Average_per_Movie],maiorMediaPorFilme))



'''O nome do(s) filme(s) mais frequente(s) e sua respectiva frequência. '''
#cria uma lista com a frequencia dos filmes
listaContadorFilmesRepetidos=[]
for i in Fisrt_Movie:
    contador=Fisrt_Movie.count(i)
    listaContadorFilmesRepetidos.append(contador)


maisFrequente = max(listaContadorFilmesRepetidos)
indexMaisFrequente = listaContadorFilmesRepetidos.index(maisFrequente)
e4.write('O filme com maior frequencia é: "{}", com {} de frequencia.\n\n'.format(Fisrt_Movie[indexMaisFrequente],maisFrequente))  
    
    
    
'''O Autor que foi mais bem pago.'''
maisBemPago = max(Total_Gross)
indexmaisBemPago = Total_Gross.index(maisBemPago)
e5.write('O ator mais bem pago é: "{}", com {} de receita.'.format(Actor[indexmaisBemPago],maisBemPago))  


f.close()
#fecha os arquivos ja com o conteudo dos exercicios
e1.close()
e2.close()
e3.close()
e4.close()
e5.close()

#abre os arquivos para ler
e1 = open("exercicioDesafio1.txt", "r")
e2 = open("exercicioDesafio2.txt", "r")
e3 = open("exercicioDesafio3.txt", "r")
e4 = open("exercicioDesafio4.txt", "r")
e5 = open("exercicioDesafio5.txt", "r")

#mostra o conteudo dos arquivos
print(e1.read())
print(e2.read())
print(e3.read())
print(e4.read())
print(e5.read())

#fecha os arquivos novamente
e1.close()
e2.close()
e3.close()
e4.close()
e5.close()
```
